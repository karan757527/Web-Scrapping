```python
import requests
import pandas as pd
import bs4

title=[]
year=[]
runtime=[]
genre=[]
rating=[]
brief=[]
director=[]
stars=[]
votes=[]
gross=[]
print("IMDB Top 1000 Movies")
main_url="https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&start=0&ref_=adv_nxt"
for m in range(1,1000,50):
    print(f"--> Fetching -> {m} to {m+49} Movies....")
    url=main_url.replace("start=0",f"start={m}")
    r=requests.get(url)
    soup=bs4.BeautifulSoup(r.text)
    x=soup.find_all('div',class_="lister-item-content")
    for i in x:
        title.append(i.find('a').text.strip())
        year.append(i.find('span',class_="lister-item-year text-muted unbold").text.strip())
        runtime.append(i.find('span',class_="runtime").text.strip())
        genre.append(i.find('span',class_="genre").text.strip())
        rating.append(float(i.find('strong').text.strip()))
        brief.append(i.find_all('p',class_="text-muted")[1].text.strip())
        director.append(i.find('p',class_="").find('a').text)
        temp=[]
        for  j in i.find('p',class_="").find_all('a')[1:]:
            temp.append(j.text.strip())
        stars.append(temp)
        z=i.find_all('span')
        if z[-2].text=="Votes:":
            votes.append(i.find_all('span')[-1].get("data-value"))
            gross.append("NA")
        elif z[-2].text=="Gross:":
            gross.append(i.find_all('span')[-1].get("data-value"))
        if z[-5].text=="Votes:":
            votes.append(i.find_all('span')[-4].get("data-value"))
    print(f"--> Fetched -> {m} to {m+49} Movies.")
print("--> Data Scrapped Succesfully.")

print("--> Creating Data Frame....")
df=pd.DataFrame(data={
    "Title":title,
    "Year":year,
    "Length":runtime,
    "Genre":genre,
    "Rating":rating,
    "Director":director,
    "Stars":stars,
    "Total Votes":votes,
    "Gross Value":gross,
    "Brief":brief,
},index=range(1,1001))

print("--> Data Frame Created Succesfully.")
print("--> Writing To CSV and HTML Files....")
df.to_csv("imdb_scrapping_karan.csv")
df.to_html("imdb_scrapping_karan.html")
print("--> Files Written.")
```
